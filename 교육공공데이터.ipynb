```python
# 전국 초등학교에 있는 통합학급 개수에 따른 학교 리스트 정리
import csv
f = open('school.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)

school_none = []
school_one = []
school_two = []
school_three = []

for row in data :
    if row[15] == '0' :
        school_none.append(str(row[4]))
    if row[15] == '1' :
        school_one.append(str(row[4]))  
    if row[15] == '2' :
        school_two.append(str(row[4]))
    if row[15] == '3' :
        school_three.append(str(row[4]))
print(school_none)
print(school_one)
print(school_two)
print(school_three)
f.close()
```

```python
# 전국 초등학교에 있는 통합학급 개수에 따른 학교를 pie 그래프로 나타내기
import csv
import matplotlib.pyplot as plt
f = open('school.csv', 'r', encoding='utf-8')
data = csv.reader(f, delimiter=',')
header = next(data)


school_none = []
school_one = []
school_two = []
school_three = []

for row in data :
    if row[15] == '0' :
        school_none.append(str(row[4]))
    if row[15] == '1' :
        school_one.append(str(row[4]))  
    if row[15] == '2' :
        school_two.append(str(row[4]))
    if row[15] == '3' :
        school_three.append(str(row[4]))
        
plt.rc('font', family = 'Malgun Gothic')
size = [len(school_none), len(school_one), len(school_two), len(school_three)]
label = ['통합학급 0개', '통합학급 1개', '통합학급 2개', '통합학급 3개']
plt.figure(dpi=300)
plt.title('전국 초등학교 통합학급 현황')
plt.axis('equal')
plt.pie(size, labels=label, autopct ='%.1f%%')
plt.legend(loc = 2)
plt.show()
```
